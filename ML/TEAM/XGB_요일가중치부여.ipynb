## **Seed 값 고정**

```python
import os
import random
import numpy as np

def seed_everything(seed):
    random.seed(seed)
    os.environ['PYTHONHASHSEED'] = str(seed)
    np.random.seed(seed)

seed_everything(42)
```

## **데이터 불러오기**

```python
import pandas as pd
```

```python
# train, test 통합 원핫 인코딩된 데이터 불러오기
data = pd.read_csv('./modeling_dataset/onehot_encoded_시,일제외.csv')

# 요일별 가중치 점수가 포함된 데이터 불러오기
days_weights = pd.read_csv('./modeling_dataset/days_weights.csv')

# 불필요한 컬럼 drop
data = data.drop(['Unnamed: 0'], axis=1)
days_weights = days_weights.drop(['Unnamed: 0'], axis=1)

# 데이터 확인
display(data)
display(days_weights)
```

## **가중치 적용하기**

```python
# 요일별 가중치 적용
for day in days_weights['요일']:
    data[f'요일_{day}'] = data[f'요일_{day}'] * days_weights.loc[days_weights['요일'] == day, '가중치'].values[0]
```

```python
# 요일별 가중치 적용 확인
display(data)
```

## **가중치를 적용시킨 데이터 활용 모델링 진행**

### **1. 데이터 분리**

```python
# 원본 데이터 train, test에 맞춰 데이터 분리위해 원본 train 데이터 가져오기
train_df = pd.read_csv('./open/train.csv', encoding='cp949')

# 원본 train 데이터셋 길이에 맞춰 전처리된 통합 데이터 분리
train_submit = data[: len(train_df)].copy()
test_submit = data[len(train_df):].copy()

# 분리한 데이터 shape 확인
print(train_submit.shape, test_submit.shape)
```

```python
import xgboost as xgb
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.metrics import mean_squared_error

X, y = train_submit, np.log2(train_df['ECLO'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

print('X_train shape : ', X_train.shape)
print('X_test shape : ', X_test.shape)
print('y_train shape : ', y_train.shape)
print('y_test shape : ', y_test.shape)
```

### **2. XGB 모델링**

```python
# GridSearch 활용해 교차 검증 진행
from sklearn.model_selection import GridSearchCV

# 탐색할 하이퍼파라미터 값들을 정의
param_grid = {
    'max_depth': [3, 5, 7, 9],
    'learning_rate': [0.01, 0.1, 0.2, 0.3],
    'n_estimators': [50, 500, 1000, 2000]
}

# XGBoost Regressor 설정
xgb_days_weights_model = xgb.XGBRegressor(objective='reg:squarederror', eval_metric='rmse')

# GridSearchCV를 사용하여 최적의 하이퍼파라미터 탐색
grid_search = GridSearchCV(estimator=xgb_days_weights_model, param_grid=param_grid, cv=5)
grid_search.fit(X_train, y_train)
```

### **3. 평가 및 예측**   

#### **RMSE**

```python
# 최적의 하이퍼파라미터 출력
print("Best Hyperparameters:", grid_search.best_params_)

# 최적의 모델에 대한 성능 출력
xgb_days_weights_best_model = grid_search.best_estimator_
test_accuracy = xgb_days_weights_best_model.score(X_test, y_test)
print("Test accuracy:", test_accuracy)

# 훈련 정확도 출력
train_accuracy = xgb_days_weights_best_model.score(X_train, y_train)
print("훈련 정확도:", train_accuracy)

# 훈련 정확도와 테스트 정확도 비교
print("훈련 vs 테스트 정확도 차이:", train_accuracy - test_accuracy)

# 테스트 세트에 대한 예측 수행
y_pred = xgb_days_weights_best_model.predict(X_test)
print('X_test predict : ', y_pred)

# 실제값과 예측값 사이의 평가지표 rmse 추출
rmse = np.sqrt(mean_squared_error(y_test, y_pred))

# 결과 출력
print('RMSE:', rmse)
```

#### **RMSLE**

```python
# 타겟변수 역변환
y_pred_convert = np.exp2(y_pred)
y_test_convert = np.exp2(y_test)

# 실제값과 예측값 사이의 평가지표 rmsle 추출
rmsle = np.sqrt(((np.log1p(y_pred_convert) - np.log1p(y_test_convert)) ** 2).mean())

# 결과 출력
print(f'Root Mean Squared Logarithmic Error (RMSLE): {rmsle}')
```

- 가중치 미적용시 모델 RMSLE : 0.45277110406736093
- 가중치를 적용한 모델이 근소하게 성능이 좋아진것을 확인할 수 있음, 더 많은 feature에 적당한 가중치를 주면 모델 성능이 더 좋아질것으로 예상
